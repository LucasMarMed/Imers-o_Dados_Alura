##Questão 1

```python
dados_limpos = dados[dados["Valor"].str.split(expand = True)[2].isnull()] #Através do .split separei e removi os imóveis possivelmente alugados.
dados_limpos[["Moeda", "Valor"]] = dados_limpos["Valor"].str.split(expand = True) #Novamento com .split usei para separar o R$ do valor numeral
dados_limpos["Valor"] = dados_limpos["Valor"].str.replace(".","").astype(float) #Utilização do .replace para redefinir a pontuação dos valores e transforma-los em tipo float
dados_limpos.head(3) #Visualização de 3 exemplos de como ficou.
```

```python
dados_limpos["Valor em Milhões"] = dados_limpos["Valor"] / 1000000 
pd.set_option("display.precision", 6) 
dados_limpos.sample(3)  #Remoção da notação cientifica.
```

```python
import seaborn as sns
sns.set()
g = sns.scatterplot(data=dados_limpos, x="Valor em Milhões", y="Metragem") 
#Através deste gráfico foi possivel constatar que existem pontos muito fora da curva, OUTLIERS.
```

```python
g = sns.scatterplot(data=dados_limpos, x="Valor em Milhões", y="Metragem") 
plt.ylim((0,3000)) #Para melhor visualização
#Já nota-se uma progressão a respeito da metragem em relação ao valor do imóvel mais compativel com a realidade. 
```

##Questão 3

```python
#Valor do m² = Preço / Metro²
dados_limpos['Preço/m²'] = dados_limpos['Valor'] / dados_limpos['Metragem']
dados_limpos.sample(1) 
```

```python
bairros_m2= dados_limpos.groupby(['Bairro'])['Preço/m²'].mean().sort_values(ascending=False) 
bairros_m2[0:15].plot.bar() #Na posição 0, temos um possível erro num imóvel de 800 mil de 10m²
plt.title("Valor médio do m² de 15 bairros caros em SP")
plt.show()
```

```python
top_bairros = dados_limpos["Bairro"].value_counts()[:10].index
#Criação da series com os 10 bairros com mais imoveis a venda.

plt.figure(figsize=(10, 8))
ax = sns.barplot(x="Bairro", y="Preço/m²", data=dados_limpos.query("Bairro in @top_bairros")) 
ax.tick_params(axis='x', rotation=45)
plt.title("Grafico de barras dos 10 bairros com mais imoveis a venda, em relação ao preço.")
plt.show()
```
