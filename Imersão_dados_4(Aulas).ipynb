```python
from sklearn.linear_model import LinearRegression #Importação do modelo de Regressão Linear de Machine Learning
from sklearn.model_selection import train_test_split #Importação do método de divisão em dados de treino e dados de teste. 

X = dados_vendas_censo[["Metragem"]]
Y = dados_vendas_censo["Valor_anuncio"]
#Váriaveis X para 'argumento/feactures' Y 'resultado/predição'
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 20)  
lr = LinearRegression() #test_size= porcentagem dos dados que ficarão como teste. random_state= definir uma 'seed'
```

```python
X_train #Criou-se os dados para treino
```

```python
lr.fit(X_train, Y_train) #'Treinou' a LR com os dados de treino
```

```python
Y_predict_test = lr.predict(X_test) #Criação da previsão dos preços com os dados testados.
Y_predict_test
```

```python
Y_test
```

```python
Y_predict_train = lr.predict(X_train) #Criação da previsão dos preços com os dados treinados.
```

```python
Y_train
```

```python
from sklearn.metrics import mean_absolute_error, r2_score


mean_absolute_error(Y_test, Y_predict_test)
```

```python
mean_absolute_error(Y_train, Y_predict_train)
```

```python
X = dados_vendas_censo[['Metragem', 'Quartos', 'Banheiros', 'Vagas','V001','V007', 'V009']] #Realização do mesmo teste, agora adicionando mais variaveis. 

Y = dados_vendas_censo["Valor_anuncio"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 20)
lr.fit(X_train, Y_train)
Y_predict_test = lr.predict(X_test)
Y_predict_train = lr.predict(X_train)
mae_test = mean_absolute_error(Y_test, Y_predict_test)
r2_test = r2_score(Y_test, Y_predict_test)
mae_train = mean_absolute_error(Y_train, Y_predict_train)
r2_train = r2_score(Y_train, Y_predict_train)
```
